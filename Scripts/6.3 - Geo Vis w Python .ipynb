# Step 7: choropleth map

```python
# Create a data frame with ave of xGF% grouped by state

average_xGF_by_state = df_10Game_MA.groupby('State')['xGF%_MA_10G'].mean()
average_xGF_by_state.head()
```

```python
# create a folium map at a high-level zoom - xGF%_MA_10G
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = states_geo, 
    data = average_xGF_by_state,
    columns = ['State', 'xGF%_MA_10G'],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "xGF% per State").add_to(map)
folium.LayerControl().add_to(map)

map
```

```python
# Save the map as an HTML file
map.save('xGF%_state_interactive_map.html')
```

```python
# Create a data frame with ave of CF% grouped by state

average_CF_by_state = df_10Game_MA.groupby('State')['CF%_MA_10G'].mean()
average_CF_by_state.head()
```

```python
# create a folium map at a high-level zoom - xGF%_MA_10G
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = states_geo, 
    data = average_xGF_by_state,
    columns = ['State', 'CF%_MA_10G'],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "CF% per State").add_to(map)
folium.LayerControl().add_to(map)

map
```

```python
# Save the map as an HTML file
map.save('CF%_state_interactive_map.html')
```
